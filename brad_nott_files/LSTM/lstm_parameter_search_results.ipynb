## Overview

This notebook parses, sorts, and displays the results of parameter tuning the single-layer Bidirectional LSTM model

```python
import collections
import pandas as pd

#f = f= open("lstm_run_results.txt","r")
f = f= open("lstm_run_results_val_loss.txt","r")

data = collections.defaultdict(list)

for line in f:

    items = line.split(',')

    #data['accuracy'].append(items[0].split(":")[1])
    data['val_loss'].append(items[0].split(":")[1])
    
    data['lr'].append(items[1].split(":")[1])
    data['batch'].append(items[2].split(":")[1])
    data['dropout'].append(items[3].split(":")[1])
    data['lstm_units'].append(items[4].split(":")[1])
```

```python
df = pd.DataFrame(data)
```

```python
df.sort_values(by='val_loss', ascending=True)
```
